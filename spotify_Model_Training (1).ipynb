# Text Cleaning/ Text Preprocessing

* Remove the Punctuation
* remove the stopwords
* stemming/Lemmatization (i am using Lemmatization which cut the word into the base word with the meaning)

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Ensure required NLTK data is downloaded
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

# Initialize WordNetLemmatizer
wnl = WordNetLemmatizer()

# Assuming npr is a DataFrame with a column "Article"
corpus = []

for i in range(len(df)):
    # Keep only letters, remove special characters
    text = re.sub('[^a-zA-Z/]', ' ', df["text"][i])

    # Convert to lowercase
    text = text.lower()

    # Split into words
    words = text.split()

    # Remove stopwords and lemmatize words
    words = [wnl.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]

    # Join words back into a single string
    clean_text = ' '.join(words)

    # Append to corpus
    corpus.append(clean_text)

print(corpus)
```

```python

df['text'] = corpus
```

applying the **TF_IDF Vectorizer & Cosine_similarity **

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidvector = TfidfVectorizer(analyzer='word',stop_words='english')
matrix = tfidvector.fit_transform(df['text'])
similarity = cosine_similarity(matrix)
```

```python
similarity[0]
```

```python
df[df['song'] == 'Absolution']
```

```python
def recommendation(song_df):
    song_df = song_df.strip().lower()  # Clean and normalize the input song name
    matching_songs = df[df['song'].str.lower() == song_df]

    if matching_songs.empty:
        print("Song not found.")
        return None  # Or return a default recommendation

    idx = matching_songs.index[0]
    distances = sorted(list(enumerate(similarity[idx])), reverse=True, key=lambda x: x[1])

    songs = []
    for i in distances[1:]:
        song_idx = i[0]
        songs.append(df.iloc[song_idx]['song'])

    return songs
```

```python
recommendation('Absolution')
```
